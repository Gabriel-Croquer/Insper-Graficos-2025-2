```python
import matplotlib.pyplot as plt

# Cria uma figura e um eixo (área de plotagem)
fig, ax = plt.subplots(figsize=(8, 4))

# Adiciona o texto no centro da figura
# fontsize: tamanho da fonte
# ha/va: alinhamento horizontal/vertical
ax.text(0.5, 0.5, 'O Palmeiras não tem mundial',
        fontsize=20,
        ha='center',
        va='center',
        color='green',  # Cor do texto
        fontweight='bold') # Estilo da fonte

# Remove os eixos (as linhas do gráfico) para um visual mais limpo
ax.axis('off')

# Exibe a imagem gerada
plt.show()
```
